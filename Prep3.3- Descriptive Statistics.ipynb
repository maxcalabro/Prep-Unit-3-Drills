```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
geometric_dist = np.random.geometric(.5, 100)

plt.hist(geometric_dist)
plt.axvline(geometric_dist.mean(), color='red')
plt.axvline(geometric_dist.mean() + geometric_dist.std(), linestyle='dashed', color='red')
plt.axvline(geometric_dist.mean() - geometric_dist.std(), linestyle='dashed', color='red')
plt.show()
```

The mean and std are not useful for the geometric distribution because it is heavily skewed to the left and tapers off in only one direction.

```python
chisquare_dist = np.random.chisquare(3, 100)

plt.hist(chisquare_dist)
plt.axvline(chisquare_dist.mean(), color='red')
plt.axvline(chisquare_dist.mean() + chisquare_dist.std(), linestyle='dashed', color='red')
plt.axvline(chisquare_dist.mean() - chisquare_dist.std(), linestyle='dashed', color='red')
plt.show()
```

Mean and std don't seem to be useful for chisquared distribution either. It is similar to the geometric distribution but tapers off more slowly as values increase.

```python
lognormal_dist = np.random.lognormal(.3, .2, 100)

plt.hist(lognormal_dist)
plt.axvline(lognormal_dist.mean(), color='red')
plt.axvline(lognormal_dist.mean() + lognormal_dist.std(), linestyle='dashed', color='red')
plt.axvline(lognormal_dist.mean() - lognormal_dist.std(), linestyle='dashed', color='red')
plt.show()
```

It appears that the lognormal distibution gets closer to the normal distribution the smaller the value of sigma, the second input. So for low sigma ( < 0.1 ), it may be useful to look at mean and std, but as sigma increases, those values become less useful.

```python
multinomial_dist = np.random.multinomial(4, [.2, .5, .8], 100)

plt.hist(multinomial_dist)
plt.axvline(multinomial_dist.mean(), color='red')
plt.axvline(multinomial_dist.mean() + multinomial_dist.std(), linestyle='dashed', color='red')
plt.axvline(multinomial_dist.mean() - multinomial_dist.std(), linestyle='dashed', color='red')
plt.show()
```

It looks like multinomial distribution is plotting a function with three separate parameter choices, 'pval'. For the intermediate choice (0.5), mean and std might be helpful, but in general these distributions are skewed and those are probably not helpful measures.

```python
rayleigh_dist = np.random.rayleigh(3, 10000)

plt.hist(rayleigh_dist)
plt.axvline(rayleigh_dist.mean(), color='red')
plt.axvline(rayleigh_dist.mean() + rayleigh_dist.std(), linestyle='dashed', color='red')
plt.axvline(rayleigh_dist.mean() - rayleigh_dist.std(), linestyle='dashed', color='red')
plt.show()
```

The Rayleigh distribution appears to be skewed in a similar way to some of the distributions above (although it's hard to see with only 100 datapoints). The peak is just below the mean, tapers quickly toward zero and slowly as values increase. The mean might be useful in some way here, but not std.

```python
pareto_dist = np.random.pareto(50, 100)

plt.hist(pareto_dist)
plt.axvline(pareto_dist.mean(), color='red')
plt.axvline(pareto_dist.mean() + pareto_dist.std(), linestyle='dashed', color='red')
plt.axvline(pareto_dist.mean() - pareto_dist.std(), linestyle='dashed', color='red')
plt.show()
```

Very similar to the geometric distribution. Pareto skews strongly toward zero and tapers as values increase. Doesn't seem like mean or std would be useful.

```python
norm1 = np.random.normal(5, .5, 10000)
norm2 = np.random.normal(10, 1, 10000)

plt.hist(norm1, color = 'blue')
plt.hist(norm2, color = 'red')
plt.show()
```

```python
norm3 = norm1 + norm2
plt.hist(norm3)
plt.axvline(norm3.mean(), color='red')
plt.axvline(norm3.mean() + norm3.std(), linestyle='dashed', color='red')
plt.axvline(norm3.mean() - norm3.std(), linestyle='dashed', color='red')
plt.show()

print([norm3.mean(), norm3.std()])
```

Adding two normal distributions gives another normal distribution as the result. The mean is the sum of the means of the original distributions, and the std is some function of the two original std (though not the sum).
